### Поиск пропущенных значений

```python
pd.DataFrame(round(df.isna().mean()*100,1)).style.background_gradient('coolwarm')
```

Пропуски есть в столбцах:
- `vehicle_type`. Можно будет заменить наиболее встречаемым значением у таких же моделей машин (хотя у некоторых моделей есть несколько вариантов кузова);
- `gear_box`. Можно заполнить по аналогии с предыдущим столбцом;
- `model`. Эти данные можно удалить, потому что невозможно предугадать что за модель имелась ввиду.
- `fuel_type`. По аналогии со столбцом vehicle_type;
- `repaired`. Можно считать, что если данных в этом столбце нет, значит ремонта не было - заполняем пропуски значениями 'no'.

```python
#обработка пропусков
df = df.dropna(subset=['model']).reset_index(drop=True)

df['repaired'] = df['repaired'].fillna('no')

df['vehicle_type'] = df['vehicle_type'].fillna(df.groupby('model')['vehicle_type'].transform(lambda x: x.value_counts().idxmax()))
df['gear_box'] = df['gear_box'].fillna(df.groupby('model')['gear_box'].transform(lambda x: x.value_counts().idxmax()))
df['fuel_type'] = df['fuel_type'].fillna(df.groupby('model')['fuel_type'].transform(lambda x: x.value_counts().idxmax()))

pd.DataFrame(round(df.isna().mean()*100,1)).style.background_gradient('coolwarm')
```

От пропусков избавились.

```python
#создание функции для построения гистограммы и диаграммы размаха для дискретных количественных признаков

def bar_box(dataset, column):
    plt.figure(figsize=(15,3))
    plt.subplot(1, 2, 1)
    plt.title(f'Гистограмма распределения "{column}"')
    sns.countplot(x=dataset[column], palette='BuPu');
    plt.xlabel(f'Значение параметра {column}')
    plt.ylabel('Частота')
    
    plt.subplot(1, 2, 2)
    plt.xlabel(f'Значения столбца {column}')
    plt.title(f'Диаграмма размаха для {column}')
    sns.boxplot(x=dataset[column], color='Indigo')
```

```python
#создание функции для построения гистограммы и диаграммы размаха для непрерывных количественных признаков

def hist_box(dataset, column):
    plt.figure(figsize=(15, 3))
    plt.subplot(1, 2, 1)
    plt.xlabel(f'Значения столбца "{column}"')
    plt.ylabel('Частота')
    plt.title(f'Гистограмма для "{column}"')
    sns.histplot(dataset[column], color='MediumPurple')
    plt.subplot(1, 2, 2)
    plt.xlabel(f'Значения столбца {column}')
    plt.title(f'Диаграмма размаха для {column}')
    sns.boxplot(x=dataset[column], color='Indigo')
```

## Обучение моделей

### Подготовка данных

Поскольку целевой признак представляет собой непрерывную числовую переменную, то стоит использовать дерево решений для задач регреcсии и модели градиентного бустинга.

```python
#разделение на тренировочную и тестовую выборки

X_train, X_test, y_train, y_test = train_test_split(df.drop(['price'], axis=1),
                                                    df['price'],
                                                    test_size = TEST_SIZE, 
                                                    random_state = RANDOM_STATE)

X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

Разделение на выборки произошло успешно.

### Подготовка Pipeline

```python
#создание списков количественных и категориальных датасета
num_col = ['reg_year', 
           'power', 
           'km']

ord_col = ['vehicle_type', 
           'gear_box', 
           'fuel_type', 
           'repaired', 
           'brand', 
           'model']
```

В качестве кодировщика для категориальных переменных будем использовать OrdinalEncoder, чтобы не плодить больше признаков, ведь некоторые столбцы содержат в себе очень много категорий.

```python
#Pipeline для кодирования категориальных признаков

ord_pipe = Pipeline(
    [('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ord',  OrdinalEncoder( 
                             handle_unknown='use_encoded_value', unknown_value=-1
                            )
      ),
     ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
    ]
)
```

```python
# Pipeline для подготовки данных

data_preprocessor = ColumnTransformer(
    [
        ('ord', ord_pipe, ord_col),
        ('num', StandardScaler(), num_col)
    ], 
    remainder='passthrough'
)
```

```python
# Итоговый Pipeline: подготовка данных и модель

pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeRegressor(random_state=RANDOM_STATE))
])
```

### Обучение дерева решений для задач регрессии

```python
#Параметры для передачи RandomizedSearchCV для поиска оптимальной модели 
#с разными Scaler и перебором гиперпараметров

param_distributions_tree = [
    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 21),
        'preprocessor__num': [StandardScaler(), 'passthrough']  
    }]
```

```python
#поиск оптимальной модели с перебором гиперпараметров

rand_tree = RandomizedSearchCV(
    pipe_final, 
    param_distributions=param_distributions_tree, 
    cv=5, 
    scoring='neg_root_mean_squared_error', 
    n_jobs=-1, 
)
```

```python
#обучение модели на тренировочной выборке
tic = time.time()
rand_tree.fit(X_train, y_train)
toc = time.time()
time_tree_train = toc - tic 
print('Время нахождения лучшей модели:\n', round(time_tree_train, 2), 'сек')
```

```python
print('Метрика RMSE для лучшей модели:\n', round(-rand_tree.best_score_, 2), 'евро') 
print('\nЛучшая модель и её параметры:\n\n', rand_tree.best_params_) 
```

```python
tic = time.time()
rand_tree.best_estimator_.fit(X_train, y_train)
toc = time.time()
time_tree_train = toc - tic 
print('Время обучения:\n', round(time_tree_train, 2), 'сек')
```

```python
tic = time.time()
pred_tree = rand_tree.best_estimator_.predict(X_train)
toc = time.time()
time_tree_pred = toc - tic 
print('Время предсказания:\n', round(time_tree_pred, 2), 'сек')
```

### Обучение CatBoostRegressor

```python
#Параметры для передачи RandomizedSearchCV для поиска оптимальной модели 
#с разными Scaler и перебором гиперпараметров

param_distributions_catboost = [    
    {
        'models': [CatBoostRegressor(random_state=RANDOM_STATE, 
                                     verbose=False, 
                                     learning_rate=0.1)],
        'models__max_depth': range(2, 5),
        'preprocessor__num': ['passthrough'] 
    }
]
```

```python
#поиск оптимальной модели с перебором гиперпараметров

rand_catboost = RandomizedSearchCV(
    pipe_final, 
    param_distributions=param_distributions_catboost, 
    cv=5, 
    scoring='neg_root_mean_squared_error', 
    n_jobs=-1
)
```

```python
#обучение модели на тренировочной выборке

tic = time.time()
rand_catboost.fit(X_train, y_train)
toc = time.time()
time_catboost_train = toc - tic 
print('Время нахождения лучшей модели:\n', round(time_catboost_train, 2), 'сек')
```

```python
print('Метрика RMSE для лучшей модели:\n', round(-rand_catboost.best_score_, 2), 'евро') 
print('\nЛучшая модель и её параметры:\n\n', rand_catboost.best_params_) 
```

```python
tic = time.time()
rand_catboost.best_estimator_.fit(X_train, y_train)
toc = time.time()
time_catboost_train = toc - tic 
print('Время обучения:\n', round(time_catboost_train, 2), 'сек')
```

```python
tic = time.time()
pred_catboost = rand_tree.best_estimator_.predict(X_train)
toc = time.time()
time_catboost_pred = toc - tic 
print('Время предсказания:\n', round(time_catboost_pred, 2), 'сек')
```

### Обучение LGBMRegressor

```python
#Параметры для передачи RandomizedSearchCV для поиска оптимальной модели 
#с разными Scaler и перебором гиперпараметров

param_distributions_lgbm = [    
    {
        'models': [LGBMRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 5),
        'models__min_samples_leaf': range(4, 6),
        'preprocessor__num': ['passthrough'] 
    }
]
```

```python
#поиск оптимальной модели с перебором гиперпараметров

rand_lgbm = RandomizedSearchCV(
    pipe_final, 
    param_distributions=param_distributions_lgbm, 
    cv=5, 
    scoring='neg_root_mean_squared_error', 
    n_jobs=-1
)
```

```python
#обучение модели на тренировочной выборке

tic = time.time()
rand_lgbm.fit(X_train, y_train)
toc = time.time()
time_lgbm_train = toc - tic 
print('Время нахождения лучшей модели:\n', round(time_lgbm_train, 2), 'сек')
```

```python
print('Метрика RMSE для лучшей модели:\n', round(-rand_lgbm.best_score_, 2), 'евро') 
print('\nЛучшая модель и её параметры:\n\n', rand_lgbm.best_params_) 
```

```python
tic = time.time()
rand_lgbm.best_estimator_.fit(X_train, y_train)
toc = time.time()
time_lgbm_train = toc - tic 
print('Время обучения:\n', round(time_lgbm_train, 2), 'сек')
```

```python
tic = time.time()
pred_lgbm = rand_lgbm.best_estimator_.predict(X_train)
toc = time.time()
time_lgbm_pred = toc - tic 
print('Время предсказания:\n', round(time_lgbm_pred, 2), 'сек')
```
